```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
#Siddharth Vadgama
#UTA ID-1001397508
```

```python
def remove_holes(image):
	im = Image.open(image)

	if 'gif' in image:
		Image.open(image).convert('RGB').save('prev.jpg')

	image = 'prev.jpg'

	input_image = cv2.imread(image, cv2.IMREAD_GRAYSCALE)

	im_out=remove_holes_helper(input_image)

	cv2.imwrite("holes.jpg", im_out)
	return im_out
```

```python
def remove_holes_helper(input_image):
	
	
	th, image_threshold = cv2.threshold(input_image, 35, 255, cv2.THRESH_BINARY)
	
	height, width = image_threshold.shape[:2]

	mask = np.zeros((height+2, width+2), np.uint8)

	cv2.imwrite("abcd.jpg",image_threshold)

	image_fill = image_threshold.copy()
		
	

	cv2.floodFill(image_fill, mask, (0,0), 255);

	im_fill_inv = cv2.bitwise_not(image_fill)

	im_out = image_threshold | im_fill_inv

	return im_out
```

```python
def soccer_analysis(my_image):
	img=cv2.imread(my_image)
	
	temp= img.copy()

	h=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

	
	lower_field=np.array([35,26,26])

	upper_field=np.array([72,255,255])
	
	field=cv2.inRange(h,lower_field,upper_field)

	field=cv2.threshold(field,30,255,cv2.THRESH_BINARY)[1]

	kk=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(10,10))

	field=cv2.morphologyEx(field,cv2.MORPH_OPEN,kk)

	
	field=remove_holes_helper(field)
	
	thres_inverse= cv2.threshold(field,32,255,cv2.THRESH_BINARY_INV)[1]
	
	lower_red_play=np.array([0,50,50])
	upper_red_play=np.array([10,255,255])
	mask_red_player=cv2.inRange(h,lower_red_play,upper_red_play)
	red_image=img.copy()

	red_image[np.where(mask_red_player==0)]=0

	threshold_red=cv2.threshold(red_image,31,255,cv2.THRESH_BINARY)[1]
	cv2.imwrite("r1.jpg",threshold_red)

	threshold_red=cv2.cvtColor(threshold_red,cv2.COLOR_BGR2GRAY)

	threshold_red=cv2.threshold(threshold_red,31,255,cv2.THRESH_BINARY)[1]
	cv2.imwrite("r2.jpg",threshold_red)

	threshold_red[np.where(mask_red_player==1)]=0

	threshold_red=threshold_red-thres_inverse
	cv2.imwrite("r3.jpg",threshold_red)
	kk=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(1,1))

	red=cv2.morphologyEx(threshold_red,cv2.MORPH_OPEN,kk)

	cv2.imwrite("red.jpg",red)

	lower_blue_play=np.array([90,181,0])

	upper_blue_play=np.array([140,255,255])

	mask_blue_player=cv2.inRange(h,lower_blue_play,upper_blue_play)
	
	blue_image=img.copy()

	blue_image[np.where(mask_blue_player==0)]=0

	threshold_blue=cv2.threshold(blue_image,31,255,cv2.THRESH_BINARY)[1]
	cv2.imwrite("1.jpg",threshold_blue)

	threshold_blue=cv2.cvtColor(threshold_blue,cv2.COLOR_BGR2GRAY)

	threshold_blue=cv2.threshold(threshold_blue,31,255,cv2.THRESH_BINARY)[1]
	cv2.imwrite("2.jpg",threshold_blue)

	threshold_blue[np.where(mask_blue_player==1)]=0

	threshold_blue=threshold_blue-thres_inverse
	cv2.imwrite("3.jpg",threshold_blue)

	
	cv2.imshow('Soccer Field',field)
	cv2.imshow('Red Team Players',red)
	cv2.imshow('Blue Team Players',threshold_blue)	
```

```python
def ocean_segmentation(image):
	image = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
	th, threshold_image2=cv2.threshold(image,122,200,cv2.THRESH_TOZERO_INV)
	
	th, threshold_image1=cv2.threshold(image,129,255,cv2.THRESH_TOZERO)

	th, threshold_image=cv2.threshold(image,190,255,cv2.THRESH_BINARY_INV)

	th, thresh_img3=cv2.threshold(threshold_image,125,255,cv2.THRESH_BINARY)
	
	th, thresh_img4=cv2.threshold(threshold_image2,1,225,cv2.THRESH_BINARY)
	
	th_filled= remove_holes_helper(threshold_image)

	th, first_output=cv2.threshold(th_filled,101,255,cv2.THRESH_BINARY_INV)

	cv2.imwrite("ocean.jpg",first_output)
	cv2.imshow('Mountain Segmant',first_output)
```

```python
def main():
	remove_holes("data/zebra_with_holes.gif")
	soccer_analysis("data/soccer_field4.jpg")
	ocean_segmentation("data/ocean2.jpg")

if __name__ == "__main__":
	main()
```
